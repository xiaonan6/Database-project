```
import pandas as pd
month_dict_leap={'1':31,'2':29,'3':31,'4':30,'5':31,'6':30,'7':31,'8':31,'9':30,'10':31,'11':30,'12':31}
month_dict_regular={'1':31,'2':30,'3':31,'4':30,'5':31,'6':30,'7':31,'8':31,'9':30,'10':31,'11':30,'12':31}
current_year = 2020
concat_df = None
for year in range(2020,current_year+1):
  if year % 4 ==0 and year % 100 !=0:
    for month in range(1,13):
      total_day = month_dict_leap[str(month)]
      if month<10:
          month = '%0*d' % (2, month)
      for days in range(1,total_day+1):
        if days<10:
          days = '%0*d' % (2, days)
        try:
          url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/"+str(month)+"-"+str(days)+"-"+str(year)+".csv"
          df = pd.read_csv(url, error_bad_lines=False)
          print(url)
          if concat_df is not None:
            concat_df = pd.concat([concat_df, df])
          else:
            concat_df=df
        except:
          print("nothing")
```

```
#creating US Cases Dataset
US_df = world_df[world_df['Country_Region'] == "US"]
#there is some county without FIPS, so we will drop them
US_df = US_df[~US_df['FIPS'].isnull()]

US_df['Recovered']=US_df['Recovered'].fillna(0)
US_df['Deaths']=US_df['Deaths'].fillna(0)

#Change data type
US_df['Confirmed'] =US_df['Confirmed'].astype(int)
US_df['Deaths'] =US_df['Deaths'].astype(int)
US_df['Recovered'] =US_df['Recovered'].astype(int)
US_df['FIPS'] = US_df['FIPS'].astype(int)

US_df["Date"] = pd.to_datetime(US_df["Date"]).dt.date
US_df
```

```
US_df=US_df.drop_duplicates(keep='first')
US_df= US_df[['FIPS','Date','Confirmed','Deaths','Recovered']]
US_df.describe()
```

## **World**

```
#creating Worldwide Cases Dataset
world_df.loc[world_df.Province_State == 'Greenland', 'Country_Region'] = "Greenland"
World_df = world_df#[world_df['Country_Region'] != "US"]

#World_df['Province_State']=World_df['Province_State'].fillna()
#World_df['Country_Region']=World_df['Country_Region'].fillna()
World_df = World_df[["Country_Region",'Date','Confirmed','Deaths','Recovered']]
World_df["Date"] = pd.to_datetime(World_df["Date"]).dt.date
World_df['Confirmed'] = World_df.groupby(['Country_Region','Date'])['Confirmed'].transform('sum')
World_df['Deaths'] = World_df.groupby(['Country_Region','Date'])['Deaths'].transform('sum')
World_df['Recovered'] = World_df.groupby(['Country_Region','Date'])['Recovered'].transform('sum')
World_df=World_df.drop_duplicates()
World_df['Confirmed'] =World_df['Confirmed'].astype(int)
World_df['Deaths'] =World_df['Deaths'].astype(int)
World_df['Recovered'] =World_df['Recovered'].astype(int)
World_df=World_df.rename(columns={"Country_Region": "Country"})
# World_df_without = World_df[World_df['Country_Region'] != "US"]
# some = world_df[world_df['Country_Region'] == "US"]
#clean=World_df
```

```
clean=World_df
clean = clean[clean['Recovered'].notnull()]
clean = clean.replace({'Country': {'Congo (Brazzaville)': 'Congo'}})
clean = clean.replace({'Country': {'Congo (Kinshasa)': 'Dem. Rep. Congo'}})
clean = clean.replace({'Country': {'Korea, South': 'South Korea'}})
clean = clean.replace({'Country': {'Dominican Republic': 'Dominican Rep.'}})
clean = clean.replace({'Country': {'Central African Republic': 'Central African Rep.'}})
clean = clean.replace({'Country': {'South Sudan': 'S. Sudan'}})
clean = clean.replace({'Country': {'Bosnia and Herzegovina': 'Bosnia and Herz.'}})
clean = clean.replace({'Country': {'Equatorial Guinea':'Eq. Guinea'}})
newtable = clean.groupby(['Country','Date'])["Confirmed", "Deaths","Recovered"].apply(lambda x : x.astype(int).sum()).reset_index()
newtable = newtable.replace({'Country': {'US': 'United States of America'}})
World_df=newtable
newtable[newtable['Country'] == 'Central African Rep.']
World_df.describe()
```

## **Region**

```
#creating Region Dataset
US_region_df = concat_df[~concat_df['FIPS'].isnull()]
US_region_df['Latitude']=US_region_df['Latitude'].fillna(US_region_df['Lat'])
US_region_df['Longitude']=US_region_df['Longitude'].fillna(US_region_df['Long_'])
US_region_df['Province_State']=US_region_df['Province_State'].fillna(US_region_df['Province/State'])
US_region_df = US_region_df[['FIPS','Admin2','Province_State','Latitude','Longitude']]
#Change data type
US_region_df ['FIPS'] =US_region_df['FIPS'].astype(int).astype(str)
US_region_df=US_region_df.drop_duplicates(subset=['FIPS'], keep='last')
US_region_df=US_region_df.rename(columns={"Admin2": "County","Province_State":"State"})
US_region_df
```
